```python
# Question: Replace all odd numbers in arr with -1 without changing arr

b = np.arange(16)

# if we done a = b then if we make any changes in "a" then changes will reflect on a

a = b.copy()
a[a%3 == 0] = -11
print(b)
print(a)
```

```python
# Question: Stack arrays a and b vertically

# Learning Both matrix should be of same size
a = np.arange(10).reshape(2,-1)
b = np.repeat(1, 10).reshape(2,-1)

np.vstack([a,b])
```
